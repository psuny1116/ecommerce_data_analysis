# 3. 카테고리 변수 처리

```python
data['gender'].nunique()
```

```python
pd.get_dummies(data, columns = ['gender'])     # gender칼럼을 gender_female과 gender_male 칼럼으로 나눔
# 이 코드에서는 drop_first = False가 기본값인데 
# columns = ['gender'] 뒤에, drop_first = True를 하면 gender_male만 남고 gender_female 사라짐
```

- 카테고리 변수마다 이 과정을 해줘야 하므로 한번에 하기 위해 for문 활용

```python
col_list = []
for i in data.columns:
    if data[i].dtype == 'O':     # object인 칼럼만 확인
        col_list.append(i)
col_list
```

```python
for i in col_list:
    print(i, data[i].nunique())     # object인 칼럼에 있는 변수 갯수 확인
```

```python
col_list = col_list[1:]     # customerID 칼럼 제외(위 결과를 보면 customerID칼럼은 전부 다르기 때문에 할 필요 없음)
col_list
```

```python
data = pd.get_dummies(data, columns = col_list, drop_first = True)     # object인 칼럼을 더미변수로 만들고 첫번째 더미는 생략
data
```

# 5. Scaling

### 1) Min-Max scaler
- min, max 이용
- 데이터 분포의 특성을 잘 반영
- 0 ~ 1 사이 값

### 2) Standard scaler
- 평균, 표준편차 이용
- 정규화

### 3) Robust scaler
- Q1, Q3 이용
- 정규화
- Standard scaler에 비해 outlier 영향 덜 받음

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler     # 스케일링 진행
```

### 1) Min-Max scaler

```python
minmax = MinMaxScaler()     # minmax scale 진행
```

```python
data.drop('customerID', axis = 1, inplace = True)     # customerID 칼럼 제거
data
```

```python
minmax.fit(data)
```

```python
scaled_data = minmax.transform(data)
```

```python
pd.DataFrame(scaled_data, columns = data.columns)
```

### 2) Standard scaler

```python
standard = StandardScaler()
```

```python
standard.fit(data.drop('Churn_Yes', axis = 1))     # 종속변수 제거하고 scaling 해야 함
```

```python
scaled_st = standard.transform(data.drop('Churn_Yes', axis = 1))
```

```python
pd.DataFrame(scaled_st, columns = data.drop('Churn_Yes', axis = 1).columns)
```

### 3) Robust scaler

```python
rob = RobustScaler()
```

```python
rob.fit(data.drop('Churn_Yes', axis = 1))     # 종속변수 제거하고 scaling 해야 함
```

```python
scaled_rob = rob.transform(data.drop('Churn_Yes', axis = 1))
```

```python
pd.DataFrame(scaled_rob, columns = data.drop('Churn_Yes', axis = 1).columns)
```

#### 최종적으로 min-max 스케일러로 데이터 저장 후 진행

```python
scaled_data = pd.DataFrame(scaled_data, columns = data.columns)
scaled_data
```

# 6. Train/Test split

```python
from sklearn.model_selection import train_test_split     # train과 test 나누기
```

```python
X = scaled_data.drop('Churn_Yes', axis = 1)
y = scaled_data['Churn_Yes']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)
```

# 7. Modeling

```python
from sklearn.neighbors import KNeighborsClassifier     # KNN 모델링
```

```python
knn = KNeighborsClassifier(n_neighbors = 10)     # n_neighbors 기본값은 5
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)     # 예측값
```

```python
pd.DataFrame({'actual_value': y_test, 'pred_value': pred}).head(30)     # 실제값과 예측값 비교
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, pred)     # 75%의 정답률
```

```python
confusion_matrix(y_test, pred)
```

# 8. 최적의 k값 구하기(위에서는 k를 임의로 10으로 지정함)

```python
error_list = []

for i in range(1, 101):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    error_list.append(accuracy_score(y_test, pred))
    
plt.figure(figsize = (20, 10))
sns.lineplot(x = range(1, 101), y = error_list, marker = 'o', markersize = 10, markerfacecolor = 'red')
```

```python
max(error_list)
```

```python
# 최대값이 몇번째 인덱스에 있는지 확인하는 방법1

error_list.index(max(error_list))     # 53은 인덱스 위치이므로 실제 값은 54가 제일 좋은 K값
```

```python
# 최대값이 몇번째 인덱스에 있는지 확인하는 방법2

np.array(error_list).argmax()     # 가장 큰 값의 위치를 찾을 때 넘파이로 바꿔주면 argmax()함수 쓸 수 있어서 편함
```

- n_neighbors는 54일 때 가장 좋은 결과를 가짐

```python
knn = KNeighborsClassifier(n_neighbors = 54)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
```

```python
accuracy_score(y_test, pred)     # 정답률이 78%로 오름
```

```python
confusion_matrix(y_test, pred)
```
